```python
#importing libraries
import pandas as pd
import numpy as np
```

```python
#loading dataset
df = pd.read_csv("CAvideos.csv")
```

```python
df.head()
```

```python
# Step: Checking Dataset Dimensions
df.shape
```

```python
# Step: Understanding Dataset Structure and Data Types

df.info()
```

```python
# Step: Checking for Missing Values

df.isnull().sum()
```

```python
# Step: Handling Missing Values in Text Columns

df["description"] = df["description"].fillna("")
df["tags"] = df["tags"].fillna("[none]")  # no tags
```

```python
# Step: Handling Missing Values in Numerical Column (Median Imputation)

df["views"] = df["views"].fillna(df["views"].median())
```

```python
# Step: Identifying and Removing Duplicate Records

df.duplicated().sum()
df = df.drop_duplicates()
```

```python
# Step: Text Cleaning and Standardization

df["title"] = df["title"].str.strip().str.lower()
df["channel_title"] = df["channel_title"].str.strip().str.lower()
```

```python
# Step: Date Parsing

df["trending_date"] = pd.to_datetime(df["trending_date"], format='%y.%d.%m')
df["publish_time"] = pd.to_datetime(df["publish_time"])
```

```python
# Step: Converting Boolean Column to Numeric Format

df["comments_disabled"] = df["comments_disabled"].astype('boolean').fillna(False).astype(int)
```

```python
# Step: Checking for Invalid Negative Values in 'views'

 (df["views"] < 0).sum()
```

```python
# Step: Removing Records with Invalid Negative Views

df = df[df["views"] >= 0]
```

```python
# Step: Removing Unnecessary Column ('thumbnail_link')

if 'thumbnail_link' in df.columns:
    df.drop('thumbnail_link', axis=1, inplace=True)
```

```python
# Step: Final Dataset Inspection and Statistical Summary

df.info()
df.describe()
```
